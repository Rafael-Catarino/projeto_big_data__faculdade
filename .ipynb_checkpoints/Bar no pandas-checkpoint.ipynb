# Repassando o código do Rafa para a planilha do bar

### Importando o pandas

```python
import pandas as pd
```

### Importando os arquivos CSVs e mostrar a tabela.

```python
df = pd.read_csv('recebimento-detalhado-20231029005807-parte-1-de-1.csv', encoding = "ISO-8859-1", sep=';', skiprows=3)
display(df)
```

### Excluindo as colunas desnecessárias.

```python
excluir_colunas = ["Status", "Descrição", "Banco", "Agência", "Conta", "Gravame", "CNPJ da instituição origem da negociação","Nome da instituição origem da negociação", "Estabelecimento", "Número do cartão", "Código da transação","Código de autorização", "NSU", "Receba Rápido", "Conversor de moedas", "Taxa Conversor de Moedas (%)", "Canal de venda", "Tipo de captura", "Resumo da operação","Tarifa", "Código da venda", "Número da máquina", "Período considerado", "Número da operação", "Número do pedido", "Quantidade de parcelas", "Valor cobrado", "Valor pendente", "Valor total","Taxa de antecipação", "Taxa de embarque", "Valor da entrada", "Número da parcela", "Número da nota fiscal","ID", "Data de pagamento na conta Cielo", "Unnamed: 46"]
df = df.drop(excluir_colunas, axis=1)
df.head()
```

### Alterando a formatação das datas.

```python
df["Data da autorização da venda"] = pd.to_datetime(df['Data da autorização da venda'], format="%d/%m/%Y")
df["Data do lançamento"] = pd.to_datetime(df['Data do lançamento'], format="%d/%m/%Y")
df["Data de pagamento"] = pd.to_datetime(df['Data de pagamento'], format="%d/%m/%Y")
display(df)
```

### Retirar o RS das colunas com valores.

```python
df["Valor bruto"] = df["Valor bruto"].str[3:]
df["Valor descontado"] = df["Valor descontado"].str[3:]
df["Valor líquido"] = df["Valor líquido"].str[3:]

df.head()
```

### Pegar as colunas com valores, trocando o ponto por nada e a vírgula, pelo ponto, evitando o erro em números como "2.223,00".

```python
df["Valor bruto"] = df["Valor bruto"].apply(lambda x: str(x).replace(".","").replace(",","."))
df["Valor descontado"] = df["Valor descontado"].apply(lambda x: str(x).replace(".","").replace(",","."))
df["Valor líquido"] = df["Valor líquido"].apply(lambda x: str(x).replace(".","").replace(",","."))

df.head()
```

### Pegar as colunas com valores e mudar para float.

```python
df["Valor bruto"] = df["Valor bruto"].astype('float')
df["Valor descontado"] = df["Valor descontado"].astype('float')
df["Valor líquido"] = df["Valor líquido"].astype('float')

df.head()
```

### Filtrando dados por Bandeiras, Formas de Pagamento e Taxas.

```python
#band_visa = df["Forma de Pagamento"].value_counts()
#display(band_visa)


# Defina a ordem desejada para as bandeiras
ordem_bandeiras = ['Visa', 'MasterCard', 'American Express', 'Elo', 'Hipercard']

# Crie uma categoria para as bandeiras com a ordem definida
df['Bandeira'] = pd.Categorical(df['Bandeira'], categories=ordem_bandeiras, ordered=True)

# Limpe as taxas removendo vírgulas e convertendo para tipo numérico
df['Taxas (%)'] = df['Taxas (%)'].str.replace(',', '').astype(float)

# Crie os DataFrames separados para "Débito à vista" e "Crédito à vista" com a coluna "Bandeira"
debito_a_vista = df[df['Forma de Pagamento'] == 'Débito à vista'][['Forma de Pagamento', 'Bandeira', 'Taxas (%)']]
credito_a_vista = df[df['Forma de Pagamento'] == 'Crédito à vista'][['Forma de Pagamento', 'Bandeira', 'Taxas (%)']]

# Exiba os DataFrames para análises
print("Débito à vista:")
print(debito_a_vista)

print("\nCrédito à vista:")
print(credito_a_vista)
```
